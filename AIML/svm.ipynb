```python
import numpy as np
import pandas as pd
from sklearn import model_selection,datasets
```

```python
dat = datasets.load_breast_cancer()
x,y,x_lab,y_lab = dat.data,dat.target,dat.feature_names,dat.target_names
y = np.where(y == 0,-1,1)
df = pd.DataFrame(data=x,columns=x_lab)
df["target"] = y
print(df["target"].unique())
X,X_t,Y,Y_t = model_selection.train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
Y = np.array(Y).reshape(-1,1)
Y_t = np.array(Y_t).reshape(-1,1)
mu = X.mean(axis=0)
stad = X.std(axis=0)
X = (X - mu) / stad
X_t = (X_t - mu) / stad
X = np.c_[np.ones((X.shape[0],1)),X]
X_t = np.c_[np.ones((X_t.shape[0],1)),X_t]
m,n = len(df["target"].unique()),X.shape[0]
EPOCH,eps = 10000,1e-8
LR,ALPHA = 0.01,0.01


class Params:
    def __init__(self,*args):
        self.LR,self.C,self.epochs = args
            
class Linear_SVM:
    def __init__(self,X,Y,LR,C,Epochs):
        self.x = X
        self.y = Y
        self.w = None
        self.hyperparams = Params(LR,C,Epochs)

    def train(self):
        m,n = self.x.shape
        self.w = np.zeros(n,1)
        for i in range(self.hyperparams.epochs):
            for j in range(m):
                if self.y[j] * (self.x[j] @ self.w) < 1:
                    grad = self.w.copy()
                    grad[1:] -= self.hyperparams.C * self.y[i] * self.x[i,1:]
                else:
                    grad = self.w.copy()
                grad[0] = 0.0
                self.w -= self.hyperparams.LR * grad

    def predict(self,X):
        if not self.w:
            self.train()
        return np.sign(X @ self.w)
    

svm = Linear_SVM(X,Y,0.01,0.1,10000)
y_pred = svm.predict(X_t)
print(np.mean(y_pred == Y_t) * 100)


    
 

        



            








```
